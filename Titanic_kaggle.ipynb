```python
train_set["Age"]=train_set[["Age","Pclass"]]
```

```python
train_set.head()
```

```python
def impute_age(cols):
    age=cols[0]
    pclass=cols[1]
    if pd.isnull(age):
        if pclass==1:
            return 38.23
        elif pclass==2:
            return 29.87
        else:
            return 25.14
    else:
        return age
```

```python
train_set["Age"]=train_set[["Age","Pclass"]].apply(impute_age,axis=1)
```

```python
sns.heatmap(train_set.isnull(),cmap="YlGnBu",yticklabels=False)
```

```python
train_set.drop("Cabin",axis=1,inplace=True)
```

```python
train_set.head()
```

```python
train_set.dropna(inplace=True)
```

```python
train_set.info()
```

```python
train_set.head()
```

```python
sex=pd.get_dummies(train_set["Sex"],drop_first=True)
embarked=pd.get_dummies(train_set["Embarked"],drop_first=True)
```

```python
train_set.drop(["Sex","Embarked","Ticket","Name"],axis=1,inplace=True)
```

```python
train_set.head()
```

```python
train_set=pd.concat([train_set,sex,embarked],axis=1)
```

```python
train_set.drop("Survived",axis=1).head()
```

```python
train_set["Survived"].head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(train_set.drop("Survived",axis=1),train_set["Survived"],test_size=0.25,random_state=45)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
accuracy=confusion_matrix(y_test,y_pred)
```

```python
print(accuracy)
```

```python
accuracy_score(y_test,y_pred)
```
